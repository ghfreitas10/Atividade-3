```python
def exibir_menu():
    print("+===================+")
    print("| MENU DE OPERAÇÕES  ")
    print("| 1 - somar          ")
    print("| 2 - subtrair       ")
    print("| 3 - multiplicar    ")
    print("| 4 - dividir        ")
    print("| 0 - sair           ")
    print("+===================+")

def soma(a, b):
    return a + b

def subtracao(a, b):
    return a - b

def multiplicacao(a, b):
    return a * b

def divisao(a, b):
    if b == 0:
        print("Erro: Divisão por zero não é permitida.")
        return None
    else:
        return a / b

while True:
    exibir_menu()
    op = input("Digite a operação desejada: ")

    if op == '0':
        print("Obrigado por usar a calculadora!")
        break

    if op not in ['1', '2', '3', '4']:
        print("Opção inválida. Por favor, escolha uma opção válida.")
        continue

    try:
        n1 = float(input("Digite o primeiro número: "))
        n2 = float(input("Digite o segundo número: "))

        if op == '1':
            print(f'O resultado da soma é: {soma(n1, n2)}')
        elif op == '2':
            print(f'O resultado da subtração é: {subtracao(n1, n2)}')
        elif op == '3':
            print(f'O resultado da multiplicação é: {multiplicacao(n1, n2)}')
        elif op == '4':
            resultado_divisao = divisao(n1, n2)
            if resultado_divisao is not None:
                print(f'O resultado da divisão é: {resultado_divisao}')
    except ValueError:
        print("Por favor, insira números válidos.")
```
